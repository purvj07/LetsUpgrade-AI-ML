```python
import pandas as pd
dataset = pd.read_csv('general_data.csv')
dataset.head()
```

```python
from sklearn import preprocessing
Label_encoder=preprocessing.LabelBinarizer()  #labael.binarizer for 2 category like 'Yes'or'No.
dataset['Attrition']=Label_encoder.fit_transform(dataset['Attrition'])  #fit_transform is used to transform categorical in to binary.
dataset.head()
```

```python
#two diff dataset of attrition_Yes and att_No

## H0: There is no significant differences in the Distance From Home between attrition (Y) and attirition (N)
##Ha: There is significant differences in the Distance From Home between attrition (Y) and attirition (N)
dataset_No = dataset[dataset['Attrition']==0]
dataset_Yes = dataset[dataset['Attrition']==1]
from scipy.stats import pearsonr
from scipy.stats import mannwhitneyu
stats,p = mannwhitneyu(dataset_No.DistanceFromHome,dataset_Yes.DistanceFromHome)
print(stats,p)
if (p<0.05):
    print('Reject null Hypothesis')
elif (p>=0.05):
    print('Null hypothesis is accepted')
```
