```python
import pandas as pd
dataset= pd.read_csv('general_data.csv')
dataset.head()
```

```python
dataset.isnull()
```

```python
dataset.dropna()
```

```python
Attrition_Yes = dataset[dataset['Attrition']=='Yes']
Attrition_Yes.describe()
```

```python
dataset.columns
```

```python
dummy = pd.get_dummies(dataset['Attrition'])
dummy.head()
```

```python
dataset1 = pd.concat((dataset,dummy),axis=1)
dataset1.head()
```

```python

dataset1.drop(['Attrition'],axis=1,inplace=True)
```

```python
dataset1.drop(['No'],axis=1,inplace=True)
dataset1.head()
```

```python
dataset1.rename(columns={"Yes":"Attrition1"},inplace=True)
```

# Hypotheis1
H0:Age has no significance over attrition

H1:Age has significance over attrition


```python
from scipy.stats import pearsonr
stats,p=pearsonr(dataset1.Attrition1,dataset1.Age)
print(stats,p)
if (p<0.05):
    print("Reject H0")
elif (p>=0.05):
    print("Null hypothesis is Accepted")
```

here the value of r is negative so both the attribute are negatively correlated and the graph will be straight line in downward direction.
